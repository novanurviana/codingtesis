```python
import matplotlib.pyplot as plt

year = ('2016', '2017', '2018', '2019', '2020','2021','2022')
year2 = ('2016', '2017', '2018', '2019', '2020')
s=(67,69,66,72,73)
y= (68,71,72,73,73,73,74)

fig,ax=plt.subplots()

ax.bar(year,y, color='y', label="Sentiment Analysis")
ax.set_ylim(40,100)
ax.plot(year2,s, label="Democracy Index", linewidth=5)
plt.gca().legend(('Democracy Index','Sentiment Analysis'))
ax.set_title('Sentiment Analysis and Democracy Index 2016-2022')
plt.ylabel("Percent")
plt.xlabel("Year")
for i in range (len(year)):
    plt.text(i,y[i],y[i], ha="center", va="bottom")
    
for j in range (len(year2)):
    plt.text(j,s[j],s[j], ha="center", va="top")
    
ax.legend()
plt.show()
```
